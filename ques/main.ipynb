```python
import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
file_name = "æµ·.csv"
```

```python
df = pd.read_csv(file_name, encoding="utf-8")
# Chinese dataframe. select the rows with 'æ¯èª' is 'ä¸­æ–‡'
df_ch = df[df['æ¯èª'] == 'ä¸­æ–‡']
# English dataframe. select the rows with 'æ¯èª' is 'è‹±æ–‡'
df_en = df[df['æ¯èª'] == 'English']
```

```python
import re

def clean_text(text, adj=False):
    text = text.replace("ã€", " ")
    text = text.replace("ï¼Œ", " ")
    text = text.replace(",", " ")
    text = text.replace("ã€‚", " ")
    # to lower
    text = text.lower()
    if adj:
        text = text.replace("çš„", " ")
    #
    pattern = r'\d+\.\s*'
    text = re.sub(pattern, '', text)
    text = text.split()
    return text
# clean df['è¯æƒ³åè©'] with function clean_text
df_ch['åè©'] = df_ch['è¯æƒ³åè©'].apply(clean_text)
df_en['åè©'] = df_en['è¯æƒ³åè©'].apply(clean_text)
df_ch['å½¢å®¹è©'] = df_ch['è¯æƒ³å½¢å®¹è©'].apply(clean_text, adj=True)
df_en['å½¢å®¹è©'] = df_en['è¯æƒ³å½¢å®¹è©'].apply(clean_text)
```

```python
from collections import Counter
# count the word frequency of df_ch['åè©'] with counter
word_freq_ch = Counter()
for text in df_ch['åè©']:
    word_freq_ch.update(text)

# count the word frequency of df_en['åè©'] with counter
word_freq_en = Counter()
for text in df_en['åè©']:
    word_freq_en.update(text)
# count the word frequency of df_ch['å½¢å®¹è©'] with counter
adj_freq_ch = Counter()
for text in df_ch['å½¢å®¹è©']:
    text = [word + "çš„" for word in text]
    adj_freq_ch.update(text)
# count the word frequency of df_en['å½¢å®¹è©'] with counter
adj_freq_en = Counter()
for text in df_en['å½¢å®¹è©']:
    adj_freq_en.update(text)
    
```

```python
# save to csv files with row name 'word' and 'freq'
pd.DataFrame(word_freq_ch.most_common(), columns=['word', 'freq']).to_csv('word_freq_ch.csv', index=False)
pd.DataFrame(word_freq_en.most_common(), columns=['word', 'freq']).to_csv('word_freq_en.csv', index=False)
pd.DataFrame(adj_freq_ch.most_common(), columns=['word', 'freq']).to_csv('adj_freq_ch.csv', index=False)
pd.DataFrame(adj_freq_en.most_common(), columns=['word', 'freq']).to_csv('adj_freq_en.csv', index=False)
```
