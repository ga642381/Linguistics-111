```python
import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
file_name = "海.csv"
```

```python
df = pd.read_csv(file_name, encoding="utf-8")
# Chinese dataframe. select the rows with '母語' is '中文'
df_ch = df[df['母語'] == '中文']
# English dataframe. select the rows with '母語' is '英文'
df_en = df[df['母語'] == 'English']
```

```python
import re

def clean_text(text, adj=False):
    text = text.replace("、", " ")
    text = text.replace("，", " ")
    text = text.replace(",", " ")
    text = text.replace("。", " ")
    # to lower
    text = text.lower()
    if adj:
        text = text.replace("的", " ")
    #
    pattern = r'\d+\.\s*'
    text = re.sub(pattern, '', text)
    text = text.split()
    return text
# clean df['聯想名詞'] with function clean_text
df_ch['名詞'] = df_ch['聯想名詞'].apply(clean_text)
df_en['名詞'] = df_en['聯想名詞'].apply(clean_text)
df_ch['形容詞'] = df_ch['聯想形容詞'].apply(clean_text, adj=True)
df_en['形容詞'] = df_en['聯想形容詞'].apply(clean_text)
```

```python
from collections import Counter
# count the word frequency of df_ch['名詞'] with counter
word_freq_ch = Counter()
for text in df_ch['名詞']:
    word_freq_ch.update(text)

# count the word frequency of df_en['名詞'] with counter
word_freq_en = Counter()
for text in df_en['名詞']:
    word_freq_en.update(text)
# count the word frequency of df_ch['形容詞'] with counter
adj_freq_ch = Counter()
for text in df_ch['形容詞']:
    text = [word + "的" for word in text]
    adj_freq_ch.update(text)
# count the word frequency of df_en['形容詞'] with counter
adj_freq_en = Counter()
for text in df_en['形容詞']:
    adj_freq_en.update(text)
    
```

```python
# save to csv files with row name 'word' and 'freq'
pd.DataFrame(word_freq_ch.most_common(), columns=['word', 'freq']).to_csv('word_freq_ch.csv', index=False)
pd.DataFrame(word_freq_en.most_common(), columns=['word', 'freq']).to_csv('word_freq_en.csv', index=False)
pd.DataFrame(adj_freq_ch.most_common(), columns=['word', 'freq']).to_csv('adj_freq_ch.csv', index=False)
pd.DataFrame(adj_freq_en.most_common(), columns=['word', 'freq']).to_csv('adj_freq_en.csv', index=False)
```
